# Init Method

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print('Constructer Method for Value Initial!....')

    def details(self):
        print(self.name, ' your age is ', self.age)
```

```python
# Create the Object

stud01 = Student('Sara', 25)
stud01.details()
```

```python
# Create the Object

stud02 = Student('Lara', 22)

stud02.details()
```

```python
stud01.details()

print('----------------------------------')

stud02.details()
```

# Property Decorator

The @prperty decorator in Python allows a method to be accessed like an attribute and is used to implement getter, setter.

# Property Getter

```python
class student:
    def __init__(self, total):
        self.value = total

    @property
    def total(self):
        return self.value
```

```python
stud01 = student(450)

print(stud01.total)
```

# Property Setter

```python
class student:
    def __init__(self, total):
        self.value = total

    @property
    def total(self):
        return self.value

    @total.setter
    def total(self, t):
        print('Setter Function Called Successfully!....')
```

```python
stud01 = student(450)

print(stud01.total)

print('----------------------------------------')

stud01.total = 20
```

# To Find Give Total is Valid or InValid

```python
class student:
    def __init__(self, total):
        self.value = total

    def average(self):
        return self.value / 5

    @property
    def total(self):
        return self.value

    @total.setter
    def total(self, t):
        if ((t < 0) or (t > 500)):
            print(t, ' - Give a Valid Total. Above 0 or Below 500')
        else:
            self.value = t
```

```python
stud01 = student(450)

print('----------------------------------------')

print(stud01.average())

print('----------------------------------------')

print(stud01.total)

print('----------------------------------------')

stud01.total = 250

print(stud01.average())
```

```python
stud02 = student(450)

print(stud02.total)

print('----------------------------------------')

stud02.total = 400

print(stud02.average())
```

```python
stud01 = student(600)

print('----------------------------------------')

print(stud01.average())

print('----------------------------------------')

print(stud01.total)

print('----------------------------------------')

stud01.total = 700

print(stud01.average())
```
