# Inheritance

# 01 Single Inheritance

```python
class Department:
    department_id = 10
    department_name = 'English'

class ClassA(Department):
    class_id = 1
    class_name  = 'English - A'

# Create the Object

stud01 = ClassA()

print(stud01.class_id)

print(stud01.class_name)

print('--------------------------------------------')

print(stud01.department_id)

print(stud01.department_name)
```

# Multiple Inheritance

```python
class Father:
    def playing(self):
        print('Father have Playing Chess!...')

    def ownBike(self):
        print('Father have Own Bike!...')

class Mother:
    def playing(self):
        print('Mother have Playing Chess!...')

    def cooking(self):
        print('Mother have Cooking Skill!...')

class Son(Father, Mother):
    def reading(self):
        print('Son have Reading Skill!...')

# Create the Object

obect01 = Son()
```

```python
obect01.reading()

print('-----------------------------------------')

obect01.playing()

print('-----------------------------------------')

obect01.cooking()

print('-----------------------------------------')

obect01.ownBike()
```

# Multi Level Inheritance

```python
class Father:
    def playing(self):
        print('Father have Playing Chess!...')

    def ownBike(self):
        print('Father have Own Bike!...')

class Mother(Father):
    def playing(self):
        print('Mother have Playing Chess!...')

    def cooking(self):
        print('Mother have Cooking Skill!...')

class Son(Mother):
    def reading(self):
        print('Son have Reading Skill!...')

# Create the Object

obect01 = Son()
```

```python
obect01.reading()

print('-----------------------------------------')

obect01.playing()

print('-----------------------------------------')

obect01.cooking()

print('-----------------------------------------')

obect01.ownBike()
```

# Diomand Problem

```python
class A:
    def display(self):
        print('I am a Class Of A')

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

object01 = D()
```

```python
object01.display()
```
