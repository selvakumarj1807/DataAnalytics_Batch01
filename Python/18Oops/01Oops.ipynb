# Class Declaration

```python
class Student:
    name = 'Sara'
    age = 25
```

# Call the Variable from the class

```python
# Dot Method

print(Student.name)

# get attribute Method

print(getattr(Student, 'age'))
```

```python
# get attribute Method

print(getattr(Student, 'gender', 'male'))
```

# Class Declaration

```python
class Student:
    pass
```

```python
# Dot Method

Student.name = 'Lara'

print(Student.name)

print('----------------------------------------')

# Set Attribute

setattr(Student, 'gender', 'male')

print(Student.gender)
```

# Remove Value From the Class

```python
class Student:
    name = 'Sara'
    age = 25
```

```python
# Dot Method

print(Student.name)

del(Student.name)

print(Student.name)
```

```python
# dell attribute

print(Student.age)

delattr(Student, 'age')

print(Student.age)
```

# Declare the Function Inside the Class

```python
class Student:
    def details(name):
        print('Hi', name)
```

# Call the Function Using the Class

```python
# Dot Method

Student.details('Sara')

# get attribute

getattr(Student, 'details')('Lara')
```

# Object

```python
class Student:
    name = 'Student Name'
    age = 'Student Age'
```

```python
# Create the Object

student01 = Student()

print(student01.name)

print(student01.age)

print('----------------------------------')

student01.name = 'Sara'
student01.age = 27

print('----------------------------------')

print(student01.name)

print(student01.age)
```

```python
# Create the Object

student02 = Student()

print(student02.name)

print(student02.age)

print('----------------------------------')

student02.name = 'Lara'
student02.age = 28

print('----------------------------------')

print(student02.name)

print(student02.age)
```

```python
print('Hi', student01.name, 'your age is', student01.age)

print('Hi', student02.name, 'your age is', student02.age)
```
