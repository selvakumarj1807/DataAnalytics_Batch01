```python
# Library Management
# Create the Class

class Library:
    def __init__(self, books):
        self.books = books

    def displayBook(self):
        print('The Books are..')
        for i in self.books:
            print(i, end=', ')

    def borrowBook(self, book):
        if book in self.books:
            self.books.remove(book)
            print('Get your', book, 'Book Now!...')
        else:
            print(book, 'Book Not in your Library!...')

    def reciveBook(self, book):
        self.books.append(book)
        print('Your Book', book, 'Received Library Successfully!...')
```

```python
# Project Development

books = ['C', 'C++', 'Java']

object01 = Library(books)

print('Library Management')

msg = '''
1. Display Books
2. Borrow Book
3. Recive Book
'''

while True:
    print(msg)

    inpt = input('Enter your Choice (1/2/3) : ')
    print('')
    
    if inpt == '1':
        object01.displayBook()
        print('')
    
    elif inpt == '2':
        book = input('Enter your Borrow book Name : ')

        object01.borrowBook(book)
        
    elif inpt == '3':
        book = input('Enter your Recive book Name : ')

        object01.reciveBook(book)
    
    else:
        print('Give a Valid Choice')
        print('Thank You Come Again!...')
        break
        
```
