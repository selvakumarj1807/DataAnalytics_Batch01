# 02 No Return Type with Arguments

```python
# Create the Function

def addTwoNum(v1, v2):
    result = v1 + v2

    print(v1, '+', v2, '=', result)
```

```python
# Call the Function

addTwoNum(10, 2)

print('---------------------------------------')

addTwoNum(20, 4)
```

```python
# Create the Function

def welcome(course):
    print('Welcome ', course, '!....')
```

```python
# Call the Function

welcome('python')

welcome('java')
```

# 03 Return Type Without Arguments

```python
# Create the Function

def addTwoNum():
    v1 = int(input('Enter your Value 01 : '))
    v2 = int(input('Enter your Value 02 : '))

    result = v1 + v2

    return result
```

```python
# Call the Function

print(addTwoNum())

print('---------------------------------------')

result01 = addTwoNum()

print('Result 01 : ', result01)
```

# 04 Return type With Arguments

```python
# Create the Function

def addTwoNum(v1, v2):
    result = v1 + v2

    return result
```

```python
# Call the Function

print(addTwoNum(10, 2))

print('---------------------------------------')

result01 = addTwoNum(20, 4)

print('Result 01 : ', result01)

print('---------------------------------------')

value1 = 30
value2 = 6

result02 = addTwoNum(value1, value2)

print('Result 02 : ', result02)
```

# Function Methods

# Arbitrary Argument Functions

```python
# Create the Function

def details(*data):
    print(data, type(data))
```

```python
# Call the Function

details(101, 'Sara', 25, 'male')
```

# Key Word Arguments Function

```python
# Create the Function

def details(name, age):
    print('Hi', name, 'your age is', age)
```

```python
# Call the Function

details('Sara', 27)

details(age = 25, name = 'Lara')
```

# Keyword Arbitrary Argument Function

```python
# Create the Function

def details(**data):
    print(data, type(data))
```

```python
# Call the Function

details(age = 25, name = 'Lara', gender = 'male', city='CHE')
```

# Default Parameter Function

```python
# Create the Function

def details(name, age, state='KL'):
    print('Hi', name, 'your age is', age, 'And your State is', state)
```

```python
# Call the Function

details('Sara', 27, 'TN')

details(age = 25, name = 'Lara')
```

# Pass List Values

```python
# Create the Function

def details(data):
    print(data, type(data))
```

```python
# Call the Function

details([101, 'Sara', 25, 'male'])
```

# Lambda Function

```python
addTwoNum = lambda a, b: a + b

print(addTwoNum(10, 2))
```
