# Exception Handling or Error Handling

# TRY & EXCEPT

```python
try:
    a = 10

    b = 2
    
    print(a / b)

except:
    print('Programm have Some Error!...')
```

# To Get Error Message

```python
try:
    a = 10

    b = 0
    
    print(a / b)

except Exception as err:
    print('Programm have Some Error!...', err)
```

# Try & Except & Else

```python
try:
    a = 10

    b = 2
    
    result = a / b

except Exception as err:
    print('Programm have Some Error!...', err)

else: 
    print(result)
```

# Try & Except & Else & Finally

```python
try:
    a = 10

    b = 0
    
    result = a / b

except Exception as err:
    print('Programm have Some Error!...', err)

else: 
    print(result)

finally:
    print('Thank you Come Again!...')
```

# IndexError

```python
try:
    ageValues = [10, 20, 30, 40]
    
    print(ageValues[6])

except Exception as e:
    print(e)
```

# ValueError

```python
try:
    age = int(input('Enter your Age : '))
    
    print(age)

except Exception as e:
    print('Enter Numeric Values Only!....')

    age = int(input('Enter your Age : '))
    
    print(age)
```

# Using Multiple Except Blocks

```python
# ValueError
# ZeroDivisionError

try:
    a = int(input('Enter Value 01 : '))

    b = int(input('Enter Value 02 : '))
    
    print(a / b)

except ValueError as e:
    print('Enter Numeric Only!...', e)

except ZeroDivisionError:
    print('Divisor Cannot be Zero!..')

except Exception as e:
    print(e)
```
