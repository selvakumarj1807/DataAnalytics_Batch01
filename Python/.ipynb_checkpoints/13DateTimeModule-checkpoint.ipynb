# Date Time

```python
import datetime as dt
```

# Fetch System Date

```python
currentDate = dt.date.today()

print(currentDate)
```

# Fetch System Date & Time

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)
```

# Declare User Date

```python
userDate = dt.date(1999,7,18)

print(userDate)
```

# Declare User Date & Time

```python
userDateTime = dt.datetime(2025,10,21,10,50,55)

print(userDateTime)
```

# Split Date & Time

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('----------------------------------')

currentDate = currentDateTime.date()

print('Current Date : ', currentDate)

currentTime = currentDateTime.time()

print('Current Time : ', currentTime)
```

# Difference Between Two Dates

```python
currentDate = dt.date.today()

print('Current Date : ', currentDate)

userDate = dt.date(1999,7,18)

print('User Date : ', userDate)

print('-----------------------------------')

DiffTwoDate = currentDate - userDate

print(DiffTwoDate)
```

# Date Time Format 

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('----------------------------------')

format01 = currentDateTime.strftime('%A')

print(format01)

format02 = currentDateTime.strftime('%B')

print(format02)

format03 = currentDateTime.strftime('%d/%b/%Y')

print(format03)

format04 = currentDateTime.strftime('%D - %B : %Y')

print(format04)
```
