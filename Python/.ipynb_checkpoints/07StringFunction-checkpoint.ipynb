# String Function

```python
msg = 'hI weLComE PYtHon'

print(msg)
```

```python
upperFormat = msg.upper()

print(upperFormat)

print(msg.lower())

print(msg.title())

print(msg.capitalize())
```

```python
msg = 'hi welcome python'

print(msg)
```

```python
print(msg.find('o'))

print(msg.find('o', 10))
```

```python
print(msg.count('o'))

print(msg.count('o', 10))
```

```python
msg = 'hi welcome python'

print(len(msg))
```

```python
# Is Upper

a = 'THIS IS UPPER'

print(a.isupper())

# Is Lower

a = 'this is lower'

print(a.islower())

# Is Alnum

a = 'selva1807'

print(a.isalnum())

# Is Alpha

a = 'Selva'

print(a.isalpha())
```

# Split

```python
UserName = 'Sara, Lara, Thara, Ram, Sam'

print(UserName, type(UserName))
```

```python
splitValues = UserName.split(', ')

print(splitValues, type(splitValues))
```

# Split Lines

```python
para = '''India, officially the Republic of India, is a vast, diverse, and ancient civilization in South Asia,
characterized by its "unity in diversity" with numerous cultures, languages, and religions coexisting.
It is the world's largest democracy, the seventh-largest country by area, and has the second-largest population. 
Bordered by the Himalayas in the north and surrounded by the Indian Ocean, Bay of Bengal, and Arabian Sea,
it boasts a varied geography. 
'''

print(para)
```

```python
splitLinesValues = para.splitlines()

print(splitLinesValues, type(splitLinesValues))
```

```python
print(splitLinesValues[2])
```

# Joins

```python
nameList = ['Sara', 'Lara', 'Thara', 'Ram', 'Sam']

print(nameList, type(nameList))
```

```python
joinNames = ' : '.join(nameList)

print(joinNames, type(joinNames))
```

# Strip

```python
# Strip

a = '  Hi  '

print(len(a))

print(len(a.strip()))

# Strip

a = '##Hi##'

print(a)

print(a.strip('#'))

# lStrip & rstrip

a = '##Hi##'

print(a.lstrip('#'))

print(a.rstrip('#'))
```
