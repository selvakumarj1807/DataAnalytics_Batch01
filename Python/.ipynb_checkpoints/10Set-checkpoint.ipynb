# Set

1. Set is a collection of data type
2. Declare - { }
3. Set are used for assign multiple values in a single variable based on non index
4. Set are Mutable(changable)
5. Used for Remove Douplicate Values

```python
userData = {10, 'Sara', 27, 'Male', 10, 27, 'Male'}

print(userData, type(userData))
```

# Add

userData = {10, 'Sara', 27, 'Male', 10, 27, 'Male'}

print(userData)

print('----------------------------------------------')

userData.add(9500912258)

print('----------------------------------------------')

print(userData)


# Update

```python
userData01 = {10, 'Sara', 27, 'Male', 10, 27, 'Male'}

userData02 = {'sara@gmail.com', 9500912258}

userData01.update(userData02)

print(userData01)
```

# Remove

```python
userData = {10, 'Sara', 27, 'Male', 10, 27, 'Male'}

print(userData)

print('---------------------------------------------------')

userData.remove('Sara')

print('---------------------------------------------------')

print(userData)

print('---------------------------------------------------')

userData.remove(9500912258)

print(userData)
```

# Discard

```python
userData = {10, 'Sara', 27, 'Male', 10, 27, 'Male'}

print(userData)

print('---------------------------------------------------')

userData.discard('Sara')

print('---------------------------------------------------')

print(userData)

print('---------------------------------------------------')

userData.discard(9500912258)

print(userData)
```

# Set Operators

# Union

```python
set01 = {1, 2, 3, 4, 5, 8}

set02 = {3, 4, 5, 6, 7}

v1 = set01.union(set02)

print(v1)
```

# Intersection

```python
set01 = {1, 2, 3, 4, 5, 8}

set02 = {3, 4, 5, 6, 7}

v1 = set01.intersection(set02)

print(v1)
```

# Symmetric_differense

```python
set01 = {1, 2, 3, 4, 5, 8}

set02 = {3, 4, 5, 6, 7}

v1 = set01.symmetric_difference(set02)

print(v1)
```
