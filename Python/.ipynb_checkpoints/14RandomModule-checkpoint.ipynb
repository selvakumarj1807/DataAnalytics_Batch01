# Random Module

```python
import random
```

# randint()

```python
randomValue = random.randint(1, 100)

print(randomValue)
```

# Choice()

```python
studentNames = ['Sara', 'Lara', 'Thara', 'Ram']

randomName = random.choice(studentNames)

print(randomName)
```

# Shuffle()

```python
numbers = [1, 2, 3, 4, 5]

random.shuffle(numbers)

print(numbers)
```

# SampleList
# sample()

```python
studentNames = ['Sara', 'Lara', 'Thara', 'Ram']

randomNames = random.sample(studentNames, 2)

print(randomNames)
```
