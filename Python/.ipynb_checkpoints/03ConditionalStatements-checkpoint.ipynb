# IF

```python
age = 25

if age >= 18:
    print(age, ' 18+ So you are Eligible for vote')
```

# If & Else

```python
age = 16

if age >= 18:
    print(age, 'is 18+ So you are Eligible for vote')

else:
    print(age, 'is 18- So you are Not Eligible for vote')
```

# If & Elif & Else

```python
m1 = int(input('Enter M1 : '))

m2 = int(input('Enter M2 : '))

m3 = int(input('Enter M3 : '))

total = m1 + m2 + m3

avg = int(total / 3)


if avg >= 100:
    print(avg, ' Invalid Average')

elif avg >= 90 and avg <= 100:
    print(avg, ' A Grade')

elif avg >= 80 and avg <= 90:
    print(avg, ' B Grade')

elif avg >= 70 and avg <= 80:
    print(avg, ' C Grade')

else:
    print(avg, ' No Grade')
```

# NestedIf

```python
m1 = int(input('Enter M1 : '))

m2 = int(input('Enter M2 : '))

m3 = int(input('Enter M3 : '))

print('')

total = m1 + m2 + m3

avg = int(total / 3)


if m1 >= 35 and m2 >= 35 and m3 >= 35:
    print('Total : ', total)
    print('Average : ', avg)
    
    if avg >= 100:
        print(avg, ' Invalid Average')

    elif avg >= 90 and avg <= 100:
        print(avg, ' A Grade')
    
    elif avg >= 80 and avg <= 90:
        print(avg, ' B Grade')
    
    elif avg >= 70 and avg <= 80:
        print(avg, ' C Grade')
    
    else:
        print(avg, ' No Grade')
else:
    if m1 <= 35:
        print(m1, ' is < 35 So you are fail!....')

    if m2 <= 35:
        print(m2, ' is < 35 So you are fail!....')

    if m3 <= 35:
        print(m3, ' is < 35 So you are fail!....')
```
