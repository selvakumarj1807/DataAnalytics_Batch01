# 1 No Return Type With Out Arguments

```python
# Function Creation

def welcome():
    print('Welcome Python Functions!.....')
```

```python
# Call the Function

welcome()

welcome()

welcome()
```

```python
# Create the Function

def addTwoNum():
    a = int(input('Enter Value 01 : '))

    b = int(input('Enter Value 02 : '))

    print(a, '+', b, '=', a + b)
```

```python
# Call the Function

addTwoNum()

print('---------------------------------------')

addTwoNum()
```

# 2.  No Return Type With Arguments

```python
# Create the Function

def addTwoNum(a, b):
    print(a, '+', b, '=', a + b)
```

```python
# Call the Function

addTwoNum(10, 2)

print('---------------------------------------')

addTwoNum(20,2)
```

# 3. Return Type With Out Arguments

```python
# Create the Function

def addTwoNum():
    a = int(input('Enter Value 01 : '))

    b = int(input('Enter Value 02 : '))

    c = a + b

    return c
```

```python
# Call the Function

result01 = addTwoNum()

print(result01)

print('---------------------------------------')

print(addTwoNum())
```

# 4. Return Type With Arguments

```python
# Create the Function

def addTwoNum(a, b):
    c = a + b

    return c
```

```python
# Call the Function

result01 = addTwoNum(10, 2)

print(result01)

print('---------------------------------------')

print(addTwoNum(20, 2))
```
