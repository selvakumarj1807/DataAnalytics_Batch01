# List

1. List is a collection of data type
2. Declare - [ ]
3. List are used for assign multiple values in a single variable based on index
4. List are Mutable(changable)

```python
userData = [10, 'Sara', 95.5, True, [15, 25, 35]]

print(userData, type(userData))
```

```python
print(userData[0], type(userData[0]))
print(userData[1], type(userData[1]))
print(userData[2], type(userData[2]))
print(userData[3], type(userData[3]))
print(userData[4], type(userData[4]))
```

```python
print(userData[4][0])
```

# Using If

```python
userData = [10, 'Sara', 95.5, True, [15, 25, 35]]

print(userData)

print('-------------------------------------------------')

if 'Sara' in userData:
    print('Yes')

else:
    print('No')
```

# Using For

```python
userData = [10, 'Sara', 95.5, True, [15, 25, 35]]

print(userData)

print('-------------------------------------------------')

for i in userData:
    print(i)
```

```python
userData = [10, 'Sara', 95.5, True, [15, 25, 35]]

print(userData)

print('-------------------------------------------------')

ageValues = userData[4]

print(ageValues)
```

# Add Values After Variable Declaration

```python
userData = [10, 'Sara', 95.5, True]

print(userData)

print('-------------------------------------------------')
```

```python
# append

userData.append(9500912258)

print(userData)
```

```python
userData01 = [10, 'Sara', 95.5, True]

print(userData01)

userData02 = ['Male', 'sara@gmail.com']

print(userData02)

print('-------------------------------------------------')
```

```python
# extend

userData01.extend(userData02)

print(userData01)
```

# Update Existing Values

```python
userData = [10, 'Sara', 95.5, True]

print(userData)

print('-------------------------------------------------')

print(userData[1])

userData[1] = 'Ram'

print(userData[1])

print(userData)
```

# Insert Values

```python
userData = [10, 'Sara', 95.5, True]

print(userData)

print('-------------------------------------------------')

userData.insert(2, 'male')

print(userData)
```

# Remove Values

```python
# Remove Based on values

userData = [10, 'Sara', 95.5, True]

print(userData)

print('-------------------------------------------------')

userData.remove(95.5)

print(userData)
```

```python
# Remove Based on Index

userData = [10, 'Sara', 95.5, True]

print(userData)

print('-------------------------------------------------')

userData.pop(3)

print(userData)
```

# Len

```python
userData = [10, 'Sara', 95.5, True]

print(len(userData))
```

# Sort

```python
ageValues = [20, 25, 18, 22, 28, 30, 16, 32]

print(ageValues)

print('-----------------------------------------------')

ageValues.sort()

print(ageValues)

print('-----------------------------------------------')

ageValues.sort(reverse=True)

print(ageValues)
```

```python

nameValues = ['Sara', 'Lara', 'Ram', 'La', 'Barani', 'A', 'sara', 'lara']

print(nameValues)

nameValues.sort()

print(nameValues)

nameValues.sort(key=len)

print(nameValues)

print('--------------------------------------------')

print(nameValues[7])

print(len(nameValues[7]))
```
