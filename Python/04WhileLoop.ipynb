```python
i = 0 # Value Initial

n = 10

while i <= n: # Value Condition
    print(i)

    i = i + 1 # Value Increment
```

# break

```python
i = 0 

n = 10

while i <= n: 
    print(i)

    if i == 5:
        i = i + 1 
        print('Before Break!...')
        break
        print('After Break!...')

    i = i + 1 
```

# Continue

```python
i = 0 

n = 10

while i <= n: 
    print(i)

    if i == 5:
        i = i + 1 
        print('Before Continue!...')
        continue
        print('After Continue!...')

    i = i + 1 
```

# Pass

```python
i = 0 

n = 10

while i <= n: 
    print(i)

    if i == 5:
        i = i + 1 
        print('Before Pass!...')
        pass
        print('After Pass!...')

    i = i + 1 
```
