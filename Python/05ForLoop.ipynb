# For Loop

```python
# Default Value Initial 0
# Default Value Increment 1

for i in range(10):
    print(i)
```

```python
# Even Number

for i in range(0,10,2):
    print(i)
```

```python
# Odd Number

for i in range(1,10,2):
    print(i)
```

```python
for i in range(1, 6):
    print(i, end=', ')
```

# Nested Loop

```python
for i in range(1, 6): # i = 3
    for j in range(i): # j = 0, i = 3
        print(i, end='')
    print('')
```

```python
for i in range(1, 6): 
    for j in range(1, i+1): 
        print(j, end='')
    print('')
```

```python
for i in range(1, 6): 
    for j in range(1, i+1): 
        print('#', end='')
    print('')
```

```python


i = 1

n = 5

while i <= n:
    j = 0
    while j < i:
        print(i, end='')
        j += 1

    print('')
    i += 1
   
```
