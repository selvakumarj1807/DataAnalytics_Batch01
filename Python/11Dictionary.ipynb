# Dictionary

1. Dictionary is a collection of data type
2. Declare - { key: value}
3. Dictionary are used for assign multiple values in a single variable based on non index, Then based on key
4. Dictionary are Mutable(changable)
5. Used for Remove Douplicate Values

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData, type(userData))
```

# Get Values

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print('-----------------------------------')

print(userData['name'])
```

# Update Value

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print('-----------------------------------')

print(userData['name'])

print('-----------------------------------')

userData['name'] = 'Priya'

print(userData)
```

# Add Values

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print('-----------------------------------')

userData.update({'gmail': 'sara@gmail.com'})

print(userData)

print('-----------------------------------')

v1 = {'mobile': 9500912258, 'state': 'TN'}

userData.update(v1)

print(userData)
```

# Delete Values

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print('-----------------------------------')

userData.pop('gender')

print(userData)
```

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print('-----------------------------------')

# Get Values Only

v1 = userData.values()

print(v1, type(v1))

print('-----------------------------------')

# Get Keys Only

v2 = userData.keys()

print(v2, type(v2))

print('-----------------------------------')

# Get Items Only

v3 = userData.items()

print(v3, type(v3))
```

# Using For Loop

```python
userData = {'name': 'Sara', 'age': 27, 'gender': 'female'}

print(userData)

print(userData['name'])

print('-----------------------------------')

for i in userData:
    print(i, ' - ', userData[i])
```

# Group of Dictionary Values

```python
students = {
    'student01' : {'name': 'Sara', 'age': 27, 'gender': 'female'}, 
    'student02' : {'name': 'Lara', 'age': 22, 'gender': 'male'}, 
}

print(students)
```

```python

student01 = students['student01']

print(student01)

student02 = students['student02']

print(student02)
```

```python
student01_name = students['student01']['name']

print(student01_name)

student02_name = students['student02']['name']

print(student02_name)
```

# Sample Program

```python
students = {
    'student01' : {'name': 'Sara', 'age': 27, 'gender': 'female'}, 
    'student02' : {'name': 'Lara', 'age': 22, 'gender': 'male'}, 
}

print(students)
```

```python
# Hi Sara, your age is 27. And your gender is female

student01_name = students['student01']['name']

student01_age = students['student01']['age']

student01_gender = students['student01']['gender']

print('Hi ', student01_name, ', your age is ', student01_age, '. And your gender is ', student01_gender)
```

# Using For Loop

```python
for i in students:
    student_name = students[i]['name']
    
    student_age = students[i]['age']
    
    student_gender = students[i]['gender']
    
    print('Hi ', student_name, ', your age is ', student_age, '. And your gender is ', student_gender)
```
