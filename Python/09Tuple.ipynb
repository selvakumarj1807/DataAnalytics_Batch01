# Tuple

1. Tuple is a collection of data type
2. Declare - ( )
3. Tuple are used for assign multiple values in a single variable based on index
4. Tuple are ImMutable(Unchangable)

```python
userData = (10, 'Sara', 27, 'Male')

print(userData, type(userData))
```

```python
print(userData[1])
```

```python
userData[1] = 'Ram'
```

# Type Casting

```python
a = 10

b = 95.6

print(a, type(a))

print(b, type(b))

print('----------------------------------')

a = float(a)

b = int(b)

print(a, type(a))

print(b, type(b))
```

```python
userData = (10, 'Sara', 27, 'Male')

print(userData, type(userData))

print('--------------------------------------------------------')

print(userData[1])

userData = list(userData)

userData[1] = 'Ram'

print(userData, type(userData))

print('--------------------------------------------------------')

userData = tuple(userData)

print(userData, type(userData))
```
