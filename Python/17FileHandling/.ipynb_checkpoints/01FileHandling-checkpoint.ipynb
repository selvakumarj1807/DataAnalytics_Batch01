# Python File Handling

# Read

```python
readFile = open('test.txt', 'r')

print(readFile.read())

readFile.close()
```

# Write

```python
readFile = open('test.txt', 'r')

print(readFile.read())

readFile.close()

print('---------------------------------------')

writeFile = open('test.txt', 'w')

writeFile.write('This is File Write Method')

writeFile.close()

print('---------------------------------------')

readFile = open('test.txt', 'r')

print(readFile.read())

readFile.close()
```

# Append

```python
readFile = open('test.txt', 'r')

print(readFile.read())

readFile.close()

print('---------------------------------------')

appendFile = open('test.txt', 'a')

appendFile.write(' This is File Appen Method')

appendFile.close()

print('---------------------------------------')

readFile = open('test.txt', 'r')

print(readFile.read())

readFile.close()
```

# ReadLines

```python
samplePara = open('para.txt', 'r')

para = samplePara.readlines()

print(para, type(para))

samplePara.close()

print('--------------------------------------')

print(para[3])

print('--------------------------------------')

for i in para:
    print(i)
```

# File Delete

```python
import os

fileName = 'sampeDelete.txt'

if os.path.exists(fileName):
    print(fileName, 'File is There!...')
    os.remove(fileName)
    print('Deleted Successfully!...')
else:
    print(fileName, 'File is not in your Folder!...')
```
